### Data visualization

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df['iyear'],hue=df['success'])
plt.xlabel('Year of Attack',fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Number of Attacks',fontsize=15)
plt.title('Year v/s Number of Attacks',size=20)
```

* from the plot above we can see the number of attacks increase each year and their success rate as well
* there was a certain dip in the count of attacks between year 1998 - 2004 and then rapid increase in year 2012 with attacks reaching above 10k per year

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['country_txt'].value_counts()[:20].index,
            y=df['country_txt'].value_counts()[:20].values,
            palette='mako')
plt.xticks(rotation=90)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.title('Number of Attacks on Countries',size=15)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['region_txt'].value_counts()[:20].index,
            y=df['region_txt'].value_counts()[:20].values,
            palette='mako')
plt.xlabel('Regions',fontsize=10)
plt.ylabel('Number of Attacks',fontsize=10)
plt.title('Number of Attacks based on Region',size=15)
plt.xticks(rotation=90)
```

```python
pd.crosstab(df['iyear'],df['region_txt']).plot(kind='area',stacked=False,figsize=(15,10))
plt.title('Terrorist Activity in Region')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
```

* from the above graphs we see that the middle-east/North-African countries suffer the most number of terrorist attacks 

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['attacktype1_txt'].value_counts()[:20].index,
            y=df['attacktype1_txt'].value_counts()[:20].values,
            palette='mako')
plt.xticks(rotation=90)
plt.xlabel('Attack Types',fontsize=15)
plt.ylabel('Number of attacks',fontsize=15)
plt.title('Number of Types of attacks on countries',size=15)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['weaptype1_txt'].value_counts()[:20].index,
            y=df['weaptype1_txt'].value_counts()[:20].values,
            palette='mako')
plt.xticks(rotation=90)
plt.xlabel('Target of Attack',fontsize=15)
plt.ylabel('Number of targets ',fontsize=15)
plt.title('Target for the Attack',size=15)
```

* Explosion/Bombing are the main weapon types that terrorist organizations uses to do the attacks 

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df['targtype1_txt'].value_counts()[:20].index,
            y=df['targtype1_txt'].value_counts()[:20].values,
            palette='mako')
plt.xticks(rotation=90)
plt.xlabel('Target of Attack',fontsize=15)
plt.ylabel('Number of targets ',fontsize=15)
plt.title('Target for the Attack',size=15)
```

* major attacks were done to target private citizens and property
